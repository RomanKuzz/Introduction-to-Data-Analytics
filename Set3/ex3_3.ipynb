```python
df.isna().sum()
```

```python
df['Fish'].fillna((df['Fish'].mean()), inplace=True)
df['Pork'].fillna((df['Pork'].mean()), inplace=True)
df['Sunflower-oil'].fillna((df['Sunflower-oil'].mean()), inplace=True)
```

```python
df[['Month', 'Year']] = df['Month'].str.split('-', expand=True)
df.head()
```

```python
df['Year'] = df['Year'].astype(int)
df.head()
```

```python
df['Year'] = df['Year'].apply(lambda x: x + 1900 if 90 <= x <= 99 else (x + 2000 if 0 <= x <= 20 else x))
df.head()
```

```python
def num_mon(month_name):
    month_map = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
                 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
    return month_map[month_name]

df['Month'] = df['Month'].apply(num_mon)
df.head()
```

```python
correlation = df.corr()
correlation.style.highlight_quantile(axis=None, q_left=0.7, color="green")
```

```python
sns.heatmap(correlation)
plt.show()
```

<b>Sunflower oil does not correlate with anything at all, I guess its price does not change much throught years</b>
