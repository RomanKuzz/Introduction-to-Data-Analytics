```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('taxis')  # Replace 'taxis' with the correct dataset name
```

```python
# Display the first few rows of the dataset to understand its structure
df.head()
```

```python
taxis_data = df.drop(['pickup_zone', 'dropoff_zone'], axis=1)
```

```python
correlation_matrix = df.corr(numeric_only=True)
correlation_matrix
```

```python
# Display a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

#Correlation Heatmap of Taxi Dataset
plt.show()
```

<h3>As we can see Fare, Distance and Total perfectly correlate with each other, tip and tolls however correlate randomly with everything else (around 0.5 with everything)</h3>
