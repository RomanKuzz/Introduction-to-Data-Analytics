```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("mobilephones.csv")
```

```python
df.head()
```

```python
sns.lmplot(data=df, x="RAM", y="Price")
plt.xlabel("RAM")
plt.ylabel("Price")
plt.show()
```

<b>RAM correlates with price pretty well, bigger memory -> higher price (which is obvious, as the chips for RAM are expensive)</b>

```python
sns.lmplot(data=df, x="RAM", y="Price", hue="Brand")
plt.xlabel("RAM")
plt.ylabel("Price")
plt.show()
```

<b>Here we see the same picture, also we can notice that apple is defenetly the most expensive model here (even though its RAM is much less than samsung ones for example, but has the same price)</b>

```python
sns.lmplot(data=df, x="BatteryCapacity", y="Price")
plt.xlabel("Battery Capacity")
plt.ylabel("Price")
plt.show()
```

<b>Battery Capacity seems to anti correlate with te price</b>

```python
sns.lmplot(data=df, x="ScreenSize", y="Price")
plt.xlabel("Screen Size")
plt.ylabel("Price")
plt.show()
```

<b>It seems that screen size nowadays does not really matter, phones have almost similiar prices with big or small sizes, but we should check it with brands specifically</b>

```python
sns.lmplot(data=df, x="ScreenSize", y="Price", hue="Brand")
plt.xlabel("Screen Size")
plt.ylabel("Price")
plt.show()
```

<b>Here it is more clear that screen size eventually does affet the price, especially if we talk about Apple, Samsung, Xiaomi or Asus
Don't really know who uses Blackberry (which has surprisingly anti correlation) or CAT nowadays (can't tell much about this one)</b>
