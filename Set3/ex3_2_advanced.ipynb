```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
survived_data = df[df['survived'] == 1]
typical_survivor = survived_data.describe()

print("Survived:")
typical_survivor
```

```python
typical_survivor.iloc[1]
```

```python
not_survived_data = df[df['survived'] == 0]
typical_non_survivor = not_survived_data.describe()

print("Not survived:")
typical_non_survivor
```

```python
typical_non_survivor.iloc[1]
```

<b>Here we can see a data of an avarage surviver and non-surviver</b>
