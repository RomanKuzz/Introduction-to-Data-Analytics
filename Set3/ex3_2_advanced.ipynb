```python
# Load the Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Display the first few rows of the dataset to understand its structure
titanic_data.head()
```

```python
# Analyze features for a typical person who survived
survived_data = titanic_data[titanic_data['survived'] == 1]
typical_survivor = survived_data.describe(include='all')

print("Survived:")
typical_survivor
```

```python
# Analyze features for a typical person who did not survive
not_survived_data = titanic_data[titanic_data['survived'] == 0]
typical_non_survivor = not_survived_data.describe(include='all')

print("Not survived:")
typical_non_survivor
```
