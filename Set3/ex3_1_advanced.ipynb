```python
import seaborn as sns
import pandas as pd
```

```python
# Load the dataset
df = sns.load_dataset('penguins')

# Create a boxenplot
sns.boxenplot(x="species", y="bill_length_mm", data=df)
```

```python
# Create a dendrogram
sns.clustermap(df.corr(numeric_only=True))
```

```python
# Create a displot
sns.displot(df, x="bill_length_mm", kde=True)
```

```python
# Create a catplot
sns.catplot(x="species", y="bill_depth_mm", hue="sex", kind="box", data=df)
```

```python
# Create a relplot
sns.relplot(x="bill_length_mm", y="body_mass_g", hue="species", data=df)
```
