```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('penguins')

# Createdd a boxenplot
sns.boxenplot(x="species", y="body_mass_g", data=df)
```

```python
df.columns
```

```python
# Created a displot
sns.displot(df, x="bill_length_mm", kde=True)
```

```python
# Created a catplot
sns.catplot(x="species", y="flipper_length_mm", hue="sex", kind="box", data=df)
```

```python
# Created a relplot
sns.relplot(x="bill_length_mm", y="body_mass_g", hue="species", data=df)
```
