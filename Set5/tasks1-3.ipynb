```python
response = requests.get("https://edu.frostbit.fi/api/events/en")
data = response.json()

# Normalize the data and extract relevant columns
df = pd.json_normalize(data)

# Extract the categories and address information
df_categories = pd.DataFrame(data[0]["categories"])
address_info = data[0]["address"]
df_address = pd.DataFrame([address_info])

# Extract the name, date and categories information
df_structure = df[['name', 'date', 'categories']]

# Combine the dataframes
df = pd.concat([df_structure, df_categories, df_address], axis=1)

# Drop the categories column
df = df.drop(columns=['categories'])

# head
df.head()
```
